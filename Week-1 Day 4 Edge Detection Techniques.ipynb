```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Example 6x6 image matrix
image = np.array([[10, 20, 30, 40, 50, 60],
                  [20, 30, 40, 50, 60, 70],
                  [30, 40, 50, 60, 70, 80],
                  [40, 50, 60, 70, 80, 90],
                  [50, 60, 70, 80, 90, 100],
                  [60, 70, 80, 90, 100, 110]], dtype=np.uint8)

# Apply Sobel operator to calculate the gradient in the x and y directions
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)  # Gradient in x direction (horizontal)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)  # Gradient in y direction (vertical)

# Calculate the gradient magnitude
magnitude = np.sqrt(sobel_x**2 + sobel_y**2)

# Calculate the gradient direction (optional)
direction = np.arctan2(sobel_y, sobel_x)

# Display the results
plt.figure(figsize=(12, 6))

# Plot the original image
plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Plot the gradient magnitude
plt.subplot(1, 3, 2)
plt.imshow(magnitude, cmap='gray')
plt.title('Gradient Magnitude')
plt.axis('off')

# Plot the gradient direction
plt.subplot(1, 3, 3)
plt.imshow(direction, cmap='hsv')
plt.title('Gradient Direction')
plt.axis('off')

plt.show()

# Print the computed results for clarity
print("Sobel X (Horizontal Gradient):\n", sobel_x)
print("Sobel Y (Vertical Gradient):\n", sobel_y)
print("Gradient Magnitude:\n", magnitude)
print("Gradient Direction:\n", direction)
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Load a grayscale image
image = cv2.imread(r'C:\Infosys\Lenna_(test_image).png', cv2.IMREAD_GRAYSCALE)

# Apply Sobel filters to get Gx and Gy
Gx = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
Gy = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)

# Compute gradient magnitude
G = np.sqrt(Gx**2 + Gy**2)

# Apply a threshold
T = 200  # Example threshold value
G_thresholded = (G >= T).astype(np.uint8) * 255  # Convert to binary image

# Display the thresholded result
plt.imshow(G_thresholded, cmap='gray')
plt.title('Thresholded Sobel Edge Detection')
plt.axis('off')
plt.show()
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Example 6x6 image matrix (grayscale)
image = np.array([[10, 20, 30, 40, 50, 60],
                  [20, 30, 40, 50, 60, 70],
                  [30, 40, 50, 60, 70, 80],
                  [40, 50, 60, 70, 80, 90],
                  [50, 60, 70, 80, 90, 100],
                  [60, 70, 80, 90, 100, 110]], dtype=np.uint8)

# Prewitt kernels for edge detection
prewitt_x = np.array([[-1, 0, 1],
                      [-1, 0, 1],
                      [-1, 0, 1]])

prewitt_y = np.array([[-1, -1, -1],
                      [ 0,  0,  0],
                      [ 1,  1,  1]])

# Apply convolution for X and Y direction
gradient_x = cv2.filter2D(image.astype(np.float64), -1, prewitt_x)
gradient_y = cv2.filter2D(image.astype(np.float64), -1, prewitt_y)

# Calculate gradient magnitude
magnitude = np.sqrt(gradient_x**2 + gradient_y**2)

# Display the results
plt.figure(figsize=(12, 6))

# Plot the original image
plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Plot the gradient magnitude (edge detection)
plt.subplot(1, 3, 2)
plt.imshow(magnitude, cmap='gray')
plt.title('Edge Detected Image (Prewitt)')
plt.axis('off')

# Plot gradient in X direction
plt.subplot(1, 3, 3)
plt.imshow(gradient_x, cmap='gray')
plt.title('Gradient in X direction')
plt.axis('off')

plt.show()

# Print results
print("Gradient in X direction:\n", gradient_x)
print("Gradient in Y direction:\n", gradient_y)
print("Gradient Magnitude:\n", magnitude)
```

```python
import cv2
import numpy as np
import streamlit as st
from matplotlib import pyplot as plt

# Load an example image (ensure it's in the same directory or provide path)
image = cv2.imread(r'C:\Infosys\Lenna_(test_image).png', cv2.IMREAD_GRAYSCALE)

# Streamlit widgets for sliders
kernel_size = st.slider('Kernel Size', 1, 10, 3, step=1)
lower_threshold = st.slider('Lower Threshold', 50, 200, 100)
upper_threshold = st.slider('Upper Threshold', 100, 300, 200)

# Apply Gaussian Blur with the selected kernel size
kernel_size = kernel_size * 2 + 1  # Ensure odd number for the kernel size
blurred_image = cv2.GaussianBlur(image, (kernel_size, kernel_size), 1.4)

# Apply Canny edge detection with selected thresholds
edges = cv2.Canny(blurred_image, lower_threshold, upper_threshold)

# Display the original image and edge-detected image using Matplotlib
fig, (ax_original, ax_edge) = plt.subplots(1, 2, figsize=(12, 6))

# Original Image
ax_original.imshow(image, cmap='gray')
ax_original.set_title('Original Image')
ax_original.axis('off')

# Edge-detected Image
ax_edge.imshow(edges, cmap='gray')
ax_edge.set_title('Canny Edge Detected Image')
ax_edge.axis('off')

# Show the plot in Streamlit
st.pyplot(fig)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the Gx and Gy matrices
Gx = np.array([[  1,   2,  -1],
                [  0,   3,   1],
                [ -1,   0,   2]])

Gy = np.array([[  0,   1,   0],
                [ -1,   0,   2],
                [  1,  -1,   0]])

# Calculate the gradient direction
direction = np.arctan2(Gy, Gx)  # Gives angle in radians

# Create a meshgrid for the arrows
Y, X = np.mgrid[0:3, 0:3]  # Create a grid for arrow placement

# Plotting
plt.figure(figsize=(8, 6))
plt.quiver(X, Y, Gx, Gy, angles='xy', scale_units='xy', scale=1, color='r')
plt.xlim(-0.5, 3)
plt.ylim(-0.5, 3)
plt.grid()
plt.title('Gradient Direction Visualization')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xticks(np.arange(0, 3, 1))
plt.yticks(np.arange(0, 3, 1))
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
